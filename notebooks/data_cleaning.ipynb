```python
# --------------------
# SETUP DRIVE
# --------------------
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Proyecto-IA-DeteccionDeFraudeEnTransacciones

# --------------------
# IMPORTS
# --------------------
import pandas as pd
import os
import numpy as np
import sys
sys.path.append('/content/drive/MyDrive/Proyecto-IA-DeteccionDeFraudeEnTransacciones/src')
from utils import load_data

# --------------------
# CARGAR DATA
# --------------------
raw_path = "data/creditcard.csv"
df = load_data(raw_path)

# --------------------
# EXPLORACIÓN BÁSICA
# --------------------
print("Dimensiones:", df.shape)
print("Nulos por columna:\n", df.isnull().sum())
print("Duplicados:", df.duplicated().sum())

# --------------------
# LIMPIEZA
# --------------------
# Eliminar duplicados
df = df.drop_duplicates()

# Llenar nulos si existieran (en este dataset no hay, pero por si acaso)
df = df.fillna(0)

# --------------------
# FEATURE ENGINEERING BÁSICO
# --------------------
# Crear log de Amount
df['Amount_log'] = np.log1p(df['Amount'])

# --------------------
# GUARDAR DATA LIMPIA
# --------------------
processed_dir = "data/processed"
os.makedirs(processed_dir, exist_ok=True)
clean_path = os.path.join(processed_dir, "creditcard_clean.csv")
df.to_csv(clean_path, index=False)
print(f"Dataset limpio guardado en: {clean_path}")
```
